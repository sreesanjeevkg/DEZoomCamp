```python
# We need to provide a name for the table; we will use 'yellow_taxi_data'
print(pd.io.sql.get_schema(taxisDF, name='greenTaxiData'))
```

```python
taxisDF.lpep_pickup_datetime = pd.to_datetime(taxisDF.lpep_pickup_datetime)
taxisDF.lpep_dropoff_datetime = pd.to_datetime(taxisDF.lpep_dropoff_datetime)
print(pd.io.sql.get_schema(taxisDF, name='greenTaxiData'))
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5432/nyTaxi')
```

```python
engine.connect()
```

```python
print(pd.io.sql.get_schema(taxisDF, name='greenTaxiData', con=engine))
```

```python
dfIter = pd.read_csv('green_tripdata_2019-09.csv', iterator=True, chunksize=100000)
dfIter
```

```python
df = next(dfIter)
df
```

```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
df
```

```python
df.head(n=0).to_sql(name='greenTaxiData', con=engine, if_exists='replace')
```

```python
%time df.to_sql(name='greenTaxiData', con=engine, if_exists='append')
```

```python
from time import time

while True: 
    try:
        t_start = time()
        df = next(dfIter)

        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        
        df.to_sql(name='greenTaxiData', con=engine, if_exists='append')

        t_end = time()

        print('inserted another chunk, took %.3f second' % (t_end - t_start))
    except StopIteration:
        print('completed')
        break
```
